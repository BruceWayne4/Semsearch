```python
import matplotlib.pyplot as plt
import numpy as np

coal = [872, 897, 943, 982, 1025]
oil = [275, 275, 285, 288, 289]
natural_gas = [45, 50, 58, 63, 64]
hydroelectric = [123, 127, 126, 132, 134]
nuclear = [3, 3, 3, 3, 3]
renewables = [57, 85, 123, 160, 204]


years = [2016, 2017, 2018, 2019, 2020]


fig, ax = plt.subplots()

ax.bar(years, coal, label='Coal', color='black')
ax.bar(years, oil, bottom=coal, label='Oil', color='red')
ax.bar(years, natural_gas, bottom=np.array(coal)+np.array(oil),
       label='Natural Gas', color='yellow')
ax.bar(years, hydroelectric, bottom=np.array(coal)+np.array(oil)+np.array(natural_gas),
       label='Hydroelectric', color='blue')
ax.bar(years, nuclear, bottom=np.array(coal)+np.array(oil)+np.array(natural_gas)+np.array(hydroelectric),
       label='Nuclear', color='green')
ax.bar(years, renewables, bottom=np.array(coal)+np.array(oil)+np.array(natural_gas)+np.array(hydroelectric)+np.array(nuclear),
       label='Renewables', color='orange')


ax.set_xlabel('Year')
ax.set_ylabel('Energy Usage (GWh)')
ax.set_xlim(2016, 2020)
ax.set_ylim(0, 2500)


ax.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt


gdp_growth = [8.5, 6.6, 7.7, 5.5, 7.5, 6.4, 7.0, 8.3, 6.8, 4.2]
pop_growth = [1.42, 1.38, 1.34, 1.30, 1.26, 1.23, 1.20, 1.17, 1.15, 1.12]
poverty_rate = [29.8, 29.0, 28.5, 27.5, 26.0, 24.9, 23.7, 22.3, 21.2, 20.0]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]


fig, ax = plt.subplots()
ax.scatter(gdp_growth, pop_growth, s=[x * 50 for x in poverty_rate], alpha=0.5)


for i in range(len(years)):
    ax.annotate(years[i], (gdp_growth[i], pop_growth[i]))


ax.set_xlabel('GDP Growth Rate (%)')
ax.set_ylabel('Population Growth Rate (%)')
ax.set_title('Bubble Chart of Macroeconomic Indicators for India (2010-2019)')


plt.legend(['Bubble size: % of population below poverty line'], loc='lower right')


plt.show()
```
